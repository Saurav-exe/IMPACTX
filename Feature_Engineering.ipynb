```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/IMPACTX/preprocessed_data.csv')
display(df.head())
```

```python

# Box plot: Modal Price by Commodity
plt.figure(figsize=(12, 6))
sns.boxplot(x='Commodity', y='Modal Price', data=df)
plt.xlabel('Commodity')
plt.ylabel('Modal Price')
plt.title('Modal Price by Commodity')
plt.xticks(rotation=90)
plt.show()

# Box plot: Modal Price by Variety
plt.figure(figsize=(12, 6))
sns.boxplot(x='Variety', y='Modal Price', data=df)
plt.xlabel('Variety')
plt.ylabel('Modal Price')
plt.title('Modal Price by Variety')
plt.xticks(rotation=90)
plt.show()

# Box plot: Modal Price by Grade
plt.figure(figsize=(12, 6))
sns.boxplot(x='Grade', y='Modal Price', data=df)
plt.xlabel('Grade')
plt.ylabel('Modal Price')
plt.title('Modal Price by Grade')
plt.show()
```

```python
sns.pairplot(df[['Min Price', 'Max Price', 'Modal Price']])
plt.suptitle('Pair Plot of Min Price, Max Price, and Modal Price')
plt.show()
```
