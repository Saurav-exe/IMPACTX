```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

##TASK: Data Loading.  

```python
data="/content/drive/MyDrive/Colab Notebooks/IMPACTX/preprocessed_data.csv"
df=pd.read_csv(data)
df.head()
```

#Task:Data Exploration.

```python

print("Shape of the DataFrame:", df.shape)

print("\nData types of columns:\n", df.dtypes)

print("\nMissing values:\n", df.isnull().sum())

print("\nDescriptive statistics for numerical variables:\n", df.describe())
```

###DATA CORRELATION:

```python
df.hist(figsize=(12, 10), bins=20)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Min Price', y='Modal Price', data=df)
plt.title('Relationship between Min Price and Modal Price')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Max Price', y='Modal Price', data=df)
plt.title('Relationship between Max Price and Modal Price')
plt.show()
```

Correlation Matrix: Heatmap representation.

```python
correlation_matrix = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

#Descriptive Statistics.

```python
# Calculate descriptive statistics for numerical variables
numerical_stats = df[['Min Price', 'Max Price', 'Modal Price']].describe()
print("Descriptive Statistics for Numerical Variables:\n", numerical_stats)

# Analyze the correlation between modal price and other numerical variables
correlation_matrix = df[['Min Price', 'Max Price', 'Modal Price']].corr()
print("\nCorrelation Matrix:\n", correlation_matrix)
```

```python
# Group by
#'Commodity' and modal price
avg_modal_price_commodity = df.groupby('Commodity')['Modal Price'].mean()
print("\nAverage Modal Price by Commodity:\n", avg_modal_price_commodity)

# 'Variety' modal price
avg_modal_price_variety = df.groupby('Variety')['Modal Price'].mean()
print("\nAverage Modal Price by Variety:\n", avg_modal_price_variety)

#'Grade' modal price
avg_modal_price_grade = df.groupby('Grade')['Modal Price'].mean()
print("\nAverage Modal Price by Grade:\n", avg_modal_price_grade)

#'StateDistrictMarket'  modal price
avg_modal_price_market = df.groupby('StateDistrictMarket')['Modal Price'].mean()
print("\nAverage Modal Price by StateDistrictMarket:\n", avg_modal_price_market)
```
